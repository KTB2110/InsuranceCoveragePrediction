```python

import pandas as pd
from sklearn.model_selection import train_test_split

def split_by_hospital(df, test_size=0.2, random_state=42):
    """
    Perform train-test split based on unique hospitals to prevent data leakage
    """
    # Get unique hospital names
    unique_hospitals = df['name'].unique()

    # Split the hospitals into train and test sets
    train_hospitals, test_hospitals = train_test_split(
        unique_hospitals,
        test_size=test_size,
        random_state=random_state
    )

    # Create train and test dataframes
    train_df = df[df['name'].isin(train_hospitals)].copy()
    test_df = df[df['name'].isin(test_hospitals)].copy()

    # Print split information
    print(f"Number of unique hospitals in training: {len(train_hospitals)}")
    print(f"Number of unique hospitals in test: {len(test_hospitals)}")
    print(f"\nTraining set size: {len(train_df)} ({(len(train_df)/len(df))*100:.1f}%)")
    print(f"Test set size: {len(test_df)} ({(len(test_df)/len(df))*100:.1f}%)")

    # Verify no leakage
    overlap = set(train_df['name'].unique()) & set(test_df['name'].unique())
    print(f"\nNumber of overlapping hospitals: {len(overlap)}")

    return train_df, test_df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

class SafeLabelEncoder:
    def __init__(self, unknown_value=-1):
        self.label_encoder = LabelEncoder()
        self.unknown_value = unknown_value
        self.known_labels = None

    def fit(self, labels):
        self.label_encoder.fit(labels)
        self.known_labels = set(labels)
        return self

    def transform(self, labels):
        labels = pd.Series(labels)
        # Mark unknown labels
        unknown_mask = ~labels.isin(self.known_labels)
        # Transform known labels
        encoded = pd.Series(index=labels.index, dtype=int)
        encoded[~unknown_mask] = self.label_encoder.transform(labels[~unknown_mask])
        encoded[unknown_mask] = self.unknown_value
        return encoded.values

    def fit_transform(self, labels):
        self.fit(labels)
        return self.transform(labels)

class HealthcarePricePipeline:
    def __init__(self):
        # Initialize encoders and transformers
        self.city_encoder = SafeLabelEncoder()
        self.zip_encoder = SafeLabelEncoder()
        self.payer_encoder = SafeLabelEncoder()
        self.tfidf = TfidfVectorizer(max_features=100)
        self.scaler = StandardScaler()

    def preprocess_data(self, df, is_training=True):
        """
        Preprocess the dataset for modeling
        """
        # Create a copy to avoid modifying original data
        df_processed = df.copy()

        # Label Encoding for categorical variables
        if is_training:
            df_processed['city'] = self.city_encoder.fit_transform(df_processed['city'])
            df_processed['zip5'] = self.zip_encoder.fit_transform(df_processed['zip5'].astype(str))
            df_processed['payer'] = self.payer_encoder.fit_transform(df_processed['payer'])
        else:
            df_processed['city'] = self.city_encoder.transform(df_processed['city'])
            df_processed['zip5'] = self.zip_encoder.transform(df_processed['zip5'].astype(str))
            df_processed['payer'] = self.payer_encoder.transform(df_processed['payer'])

        # TF-IDF for description
        if is_training:
            description_features = self.tfidf.fit_transform(df_processed['description'])
        else:
            description_features = self.tfidf.transform(df_processed['description'])

        # Convert to dense array and create new feature columns
        desc_dense = description_features.todense()
        desc_cols = [f'desc_feature_{i}' for i in range(desc_dense.shape[1])]

        # Scale numerical features (beds)
        if is_training:
            df_processed['beds'] = self.scaler.fit_transform(df_processed[['beds']])
        else:
            df_processed['beds'] = self.scaler.transform(df_processed[['beds']])

        # Combine all features
        feature_df = pd.concat([
            df_processed[['city', 'zip5', 'beds', 'payer',
                         'BCBS', 'Medicare', 'Medicaid', 'UHC', 'Aetna', 'Cigna',
                         'Cash Price', 'Gross Charge', 'HospitalInsurance']],
            pd.DataFrame(desc_dense, columns=desc_cols)
        ], axis=1)

        return feature_df

# [Previous SafeLabelEncoder and HealthcarePricePipeline classes remain the same]

def train_evaluate_model(train_df, test_df):
    """
    Train XGBoost model and evaluate performance
    """
    # Initialize pipeline
    pipeline = HealthcarePricePipeline()

    # Preprocess training and test data
    X_train = pipeline.preprocess_data(train_df, is_training=True)
    X_test = pipeline.preprocess_data(test_df, is_training=False)

    # Get target variables
    y_train = train_df['price']
    y_test = test_df['price']

    # Initialize and train XGBoost model
    model = XGBRegressor(
        n_estimators=1000,
        learning_rate=0.1,
        max_depth=7,
        min_child_weight=1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        eval_metric='rmse'  # Move eval_metric to model parameters
    )

    # Train model
    model.fit(
        X_train,
        y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
        verbose=100
    )

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    metrics = {
        'rmse': np.sqrt(mean_squared_error(y_test, y_pred)),
        'mae': mean_absolute_error(y_test, y_pred),
        'r2': r2_score(y_test, y_pred)
    }

    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': X_train.columns,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)

    return model, pipeline, metrics, feature_importance
```
