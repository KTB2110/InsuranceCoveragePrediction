```python
filt = df[df['price'] <= 35000]
```

```python
desc = pd.read_csv('/content/drive/MyDrive/si670project/df_description_clean.csv')
```

```python
desc_unique = desc[['description', 'birch_cluster']].drop_duplicates()

# Perform the merge
merged = filt.merge(desc_unique, how='inner', on='description')

# Check the number of rows
print(f"Expected rows: 588466, Merged rows: {len(merged)}")
```

```python
len(merged), len(filt)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Calculate the mean price per cluster
mean_price_per_cluster = merged.groupby('birch_cluster')['price'].mean()

# Map the mean price to each row in the DataFrame
merged['predicted_price'] = merged['birch_cluster'].map(mean_price_per_cluster)

# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(merged['price'], merged['predicted_price']))

print(f"RMSE: {rmse}")
```

```python
merged['gross_charge_price'] = merged.apply(
    lambda row: row['price'] if row['Gross Charge'] == 1 else np.nan, axis=1
)

# Fill in the gross charge price for all rows in the same hospital-description group
merged['gross_charge_price'] = merged.groupby(['name', 'description'])['gross_charge_price'].transform('max')
```

```python
df2 = merged.drop(columns=['gross_charge_price'])
```

```python
df2 = df2[(df2['Gross Charge'] != 1) & (df2['Cash Price'] != 1)]
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Calculate the mean price per cluster
mean_price_per_cluster = df2.groupby('birch_cluster')['price'].mean()

# Map the mean price to each row in the DataFrame
df2['predicted_price'] = df2['birch_cluster'].map(mean_price_per_cluster)

# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(df2['price'], df2['predicted_price']))

print(f"RMSE: {rmse}")
```
