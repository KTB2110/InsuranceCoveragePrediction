```python
# Define the mapping
mapping = {
    "BCBS": ['blue cross', 'bcn', 'bcbs', 'bieu cross', 'blue care'],
    "Medicare": ['medicare'],
    "Medicaid": ['medicaid'],
    "UHC": ['uhc', 'united'],
    "Aetna": ['aetna'],
    "Cigna": ['cigna'],
    "Cash Price": ['cash price'],
    "Gross Charge": ['gross charge']
}

# Initialize columns for each payer
for payer_category in mapping:
    df[payer_category] = 0

# One-hot encoding
for payer_category, keywords in mapping.items():
    df[payer_category] = df['payer'].str.lower().apply(
        lambda x: 1 if any(keyword in x for keyword in keywords) else 0
    )
```

```python
df.head()
```

```python
medical_words = ['hospital', 'health', 'care', 'services', 'center', 'insurance', 'plan', 'medicare', 'advantage', 'medicaid', 'blue', 'cross']

def extract_non_medical_words(text):
    words = text.lower().split()
    return [word for word in words if word not in medical_words]

df['HospitalInsurance'] = df.apply(
    lambda row: 1 if any(word in extract_non_medical_words(row['name']) for word in extract_non_medical_words(row['payer'])) else 0,
    axis=1
)
```
