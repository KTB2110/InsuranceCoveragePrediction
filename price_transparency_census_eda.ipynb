```python
def get_single_hospital_prices(cert_num):
  cert_files = os.listdir(drive_path + "parquet")
  cert_files = [f for f in cert_files if cert_num in f]
  cert_dfs = []
  for f in cert_files:
    p_files = os.listdir(drive_path + "parquet" + "/" + f)
    folder_dfs = []
    for p in p_files:
      folder_dfs.append(pd.read_parquet(drive_path + "parquet" + "/" + f + "/" + p))
    #concatenate dfs
    folder_df = pd.concat(folder_dfs)
    cert_dfs.append(folder_df)
  ret = pd.concat(cert_dfs)
  ret["cms_certification_num"] = cert_num
  return(ret)
```

```python
import requests
import pandas as pd
# Replace 'your_census_api_key_here' with your actual Census API key
API_KEY = '4026107b0471dadb6c4e191b05da5f8abf0c89ce'
BASE_URL = 'https://api.census.gov/data/2020/acs/acs5'

# Define variables for the data to be fetched
# B19013_001E: Median Household Income
# B01001_001E: Total Population
# B01001_020E to B01001_025E and B01001_044E to B01001_049E: Population over 65
# Additional variables for health insurance, disability, unemployment, poverty, and education
VARIABLES = [
    "B19013_001E", "B01001_001E",  # Median income and total population
    "B01001_020E", "B01001_021E", "B01001_022E", "B01001_023E", "B01001_024E", "B01001_025E",  # Male 65+
    "B01001_044E", "B01001_045E", "B01001_046E", "B01001_047E", "B01001_048E", "B01001_049E",  # Female 65+
    "B27010_002E", "B27010_017E",  # Population with and without health insurance
    #"C18108_001E",  # Population with a disability
    "B23025_005E", "B23025_002E",  # Unemployed population and labor force
    "B17001_002E", "B17001_001E",  # Poverty population and total population for poverty rate
    "B15003_017E", "B15003_022E", "B15003_025E"  # Educational attainment: High school, bachelor's, and graduate degrees
]
VARIABLES_STR = ",".join(VARIABLES)

# Function to get data for all ZIP codes

def fetch_census_data():
    params = {
        "get": VARIABLES_STR,
        "for": "zip code tabulation area:*",
        "key": API_KEY
    }

    response = requests.get(BASE_URL, params=params)

    if response.status_code == 200:
        data = response.json()
        # Convert the JSON response into a pandas DataFrame
        df = pd.DataFrame(data[1:], columns=data[0])

        # Filter for ZIP codes in Michigan (beginning with 48-49 as Michigan ZCTAs)
        df = df[df["zip code tabulation area"].str.startswith(('48', '49'))]
        return df
    else:
        raise Exception(f"API request failed with status code {response.status_code}: {response.text}")
```

```python
try:
    data = fetch_census_data()
    # Rename columns for better readability
    data.rename(columns={
        "B19013_001E": "Median_Income",
        "B01001_001E": "Total_Population",
        "B01001_020E": "Male_65_66",
        "B01001_021E": "Male_67_69",
        "B01001_022E": "Male_70_74",
        "B01001_023E": "Male_75_79",
        "B01001_024E": "Male_80_84",
        "B01001_025E": "Male_85+",
        "B01001_044E": "Female_65_66",
        "B01001_045E": "Female_67_69",
        "B01001_046E": "Female_70_74",
        "B01001_047E": "Female_75_79",
        "B01001_048E": "Female_80_84",
        "B01001_049E": "Female_85+",
        "B27010_002E": "With_Health_Insurance",
        "B27010_017E": "Without_Health_Insurance",
        #"C18108_001E": "With_Disability",
        "B23025_005E": "Unemployed",
        "B23025_002E": "Labor_Force",
        "B17001_002E": "Below_Poverty",
        "B17001_001E": "Poverty_Universe",
        "B15003_017E": "High_School_Grad",
        "B15003_022E": "Bachelors_Degree",
        "B15003_025E": "Graduate_Degree",
        "zip code tabulation area": "ZIP_Code"
    }, inplace=True)

    # Convert numeric columns to appropriate data types
    numeric_columns = [
        "Median_Income", "Total_Population",
        "Male_65_66", "Male_67_69", "Male_70_74", "Male_75_79", "Male_80_84", "Male_85+",
        "Female_65_66", "Female_67_69", "Female_70_74", "Female_75_79", "Female_80_84", "Female_85+",
        "With_Health_Insurance", "Without_Health_Insurance",
        "Unemployed", "Labor_Force", "Below_Poverty", "Poverty_Universe",
        "High_School_Grad", "Bachelors_Degree", "Graduate_Degree"
    ]

    for col in numeric_columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

    # Calculate total population over 65
    data["Population_Over_65"] = data[[
        "Male_65_66", "Male_67_69", "Male_70_74", "Male_75_79", "Male_80_84", "Male_85+",
        "Female_65_66", "Female_67_69", "Female_70_74", "Female_75_79", "Female_80_84", "Female_85+"
    ]].sum(axis=1)

    # Calculate percentage of population over 65
    data["Percentage_Over_65"] = (data["Population_Over_65"] / data["Total_Population"] * 100).round(2)

    # Calculate unemployment rate
    data["Unemployment_Rate"] = (data["Unemployed"] / data["Labor_Force"] * 100).round(2)

    # Calculate poverty rate
    data["Poverty_Rate"] = (data["Below_Poverty"] / data["Poverty_Universe"] * 100).round(2)

    data["HS_rate"] = (data["High_School_Grad"] / data["Total_Population"] * 100).round(2)
    data["Bachelors_rate"] = (data["Bachelors_Degree"] / data["Total_Population"] * 100).round(2)
    data["Graduate_rate"] = (data["Graduate_Degree"] / data["Total_Population"] * 100).round(2)
    data["With_Health_Insurance_Rate"] = (data["With_Health_Insurance"] / data["Total_Population"] * 100).round(2)
    data["Without_Health_Insurance_Rate"] = (data["Without_Health_Insurance"] / data["Total_Population"] * 100).round(2)
    #data["With_Disability_Rate"] = (data["With_Disability"] / data["Total_Population"] * 100).round(2)
    data["Labor_Force_Rate"] = (data["Labor_Force"] / data["Total_Population"] * 100).round(2)
    data["Below_Poverty_Rate"] = (data["Below_Poverty"] / data["Poverty_Universe"] * 100).round(2)
    data["Unemployment_Rate"] = (data["Unemployed"] / data["Labor_Force"] * 100).round(2)





    data.drop(columns=["Male_65_66", "Male_67_69", "Male_70_74", "Male_75_79", "Male_80_84", "Male_85+",
        "Female_65_66", "Female_67_69", "Female_70_74", "Female_75_79", "Female_80_84", "Female_85+", "High_School_Grad",
                       "Bachelors_Degree", "Graduate_Degree", "With_Health_Insurance",
                       "Without_Health_Insurance", "Labor_Force",
                       'Below_Poverty', "Poverty_Universe", "Population_Over_65", "Unemployed",
                       "Labor_Force", "Total_Population"],
                        inplace=True)

    # Display the data
    print(data.head())

    # Save the data to a CSV file
    data.to_csv(drive_path + "michigan_zipcode_data.csv", index=False)
    print("Data saved to michigan_zipcode_data.csv")

except Exception as e:
    print(f"Error: {e}")
```
